```python
import os
import cv2
from ultralytics import YOLO
```

```python
# 影片抽幀
model = YOLO("D:/ML/project/code/runs/detect/trainxx4/weights/best.pt")
img_id = 11316
video_frame_rate = 30
img_per_sec = 5

for i in range(91, 92):
    frame_count = 0
    video_path = "D:/ML/project/data/human_data_3/video/" + str(i) + ".mp4"
    save_path = "D:/ML/project/data/human_data_3/"
    os.makedirs(save_path + str(i), exist_ok=True)
    os.makedirs(save_path + str(i) + "/" + "images", exist_ok=True)
    os.makedirs(save_path + str(i) + "/" + "labels", exist_ok=True)
    print("VIDEO:", i)

    VIDEO_IN = cv2.VideoCapture(video_path)

    while True:
        success, img_BGR = VIDEO_IN.read()
        if not success:
            break

        if frame_count % (video_frame_rate / img_per_sec) == 0:
            img_name = save_path + str(i) + "/" + "images" + "/" + "000" + str(img_id) + ".jpg"
            label_name = save_path + str(i) + "/" + "labels" + "/" + "000" + str(img_id) + ".txt"

            img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB) 
            results = model.predict(img_RGB, save_txt = False)
            txt_content = []
            with open(label_name, "w") as f: 
                empty_frame = True
                for result in results:
                    for box in result.boxes:
                        class_id = int(box.cls)
                        x_center, y_center, width, height = box.xywhn[0]
                        f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
                        empty_frame = False
                if empty_frame:
                    f.write("0 0.5 0.5 0.5 0.5\n")
                print("Saved:", label_name)

            cv2.imwrite(img_name, img_BGR)

            # with open(label_name, "w") as txt_file:
            #     txt_file.writelines(txt_content)
            
            # print("Saved:", img_name)
            img_id += 1

        frame_count += 1
        
    VIDEO_IN.release()
    cv2.destroyAllWindows()

```
