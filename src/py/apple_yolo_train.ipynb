```python
class Box():
    def __init__(self, cls, x1, y1, x2, y2, confid):
        self.cls = cls
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.confid = confid

    def printBox(self):
        print(self.cls, ": [", self.x1, self.y1, self.x2, self.y2, "], ", self.confid)

def iou(box1, box2):
    # function to get iou
    inter_x1 = max(box1.x1, box2.x1)
    inter_y1 = max(box1.y1, box2.y1)
    inter_x2 = min(box1.x2, box2.x2)
    inter_y2 = min(box1.y2, box2.y2)

    box1_area = abs(box1.x2 - box1.x1) * abs(box1.y2 - box1.y1)
    box2_area = abs(box2.x2 - box2.x1) * abs(box2.y2 - box2.y1)

    inter_area = abs(inter_x2 - inter_x1) * abs(inter_y2 - inter_y1)
    union_area = box1_area + box2_area - inter_area

    if union_area == 0:
        return 0

    return inter_area / union_area

def weightedMergeBoxes(queue, box, iou_threshhold = 0.8, min_confid = 0.5):
    cls_list = [0, 0, 0, 0]

    if box.confid >= min_confid:
        cls_list[box.cls] += 1
    for index, past_boxes in enumerate(queue):
        for past_box in past_boxes:
            if iou(box, past_box) > iou_threshhold:
                if past_box.confid >= min_confid:
                    cls_list[past_box.cls] += 1
    
    class_index = box.cls
    for i in range(len(cls_list)):
        if(cls_list[i] > cls_list[class_index]):
            class_index = i

    return class_index
```

```python
# show video(new)
model = YOLO(config["model"])

# video_path = "D:/ML/project/data/human_data_3/video/91.mp4"
# video_path = "C:/Users/LiangYingWu/Downloads/2.mp4"
video_path = "D:/ML/project/test/test.mp4"
VIDEO_IN = cv2.VideoCapture(video_path)

fourcc = cv2.VideoWriter_fourcc(*"XVID")
VIDEO_OUT = cv2.VideoWriter("apple.avi", fourcc, 30, (1280, 720))
# VIDEO_OUT = cv2.VideoWriter("apple.avi", fourcc, 30, (1920, 1080))

pause = False

queue = deque(maxlen=5)

time_1 = time()

while True:
    if not pause:
        success, img_BGR = VIDEO_IN.read()

        if not success:
            break
        
        img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)  
        
        results = model.predict(img_RGB)
        time_2 = time()
        fps = 1 / (time_2 - time_1)
        time_1 = time_2
        cv2.rectangle(img_BGR, (0, 0), (175, 40), (255, 255, 255), cv2.FILLED)
        cv2.putText(img_BGR, "fps: " + str(round(fps, 2)), (5, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

        new_boxes = []

        for result in results:
            if len(result.boxes.conf) != 0:
                boxes = result.boxes.xyxy.cpu().numpy().astype(int)

                for index in range(len(boxes)):
                    if result.boxes.conf[index] > 0.5:
                        x1, y1, x2, y2 = boxes[index]
                        cls = int(result.boxes.cls[index])
                        confid = round(float(result.boxes.conf[index]), 3)

                        if config["modify_result"]:
                            new_box = Box(cls, x1, y1, x2, y2, confid)
                            new_boxes.append(new_box)
                            new_cls = weightedMergeBoxes(queue=queue, box=new_box, iou_threshhold=0.8, min_confid=0.7)
                            color = colors[new_cls]

                            label_content = class_name[new_cls] + " " + str(confid)
                            if new_cls != cls:
                                label_content = class_name[cls] + " => " + label_content
                        else:
                            color = colors[cls]
                            label_content = class_name[cls] + " " + str(confid)
                        
                        cv2.rectangle(img_BGR, (x1, y1), (x2, y2), color, 2)
                        label_size = cv2.getTextSize(label_content, cv2.FONT_HERSHEY_SIMPLEX, 1, 2)
                        cv2.rectangle(img_BGR, (x1, y1), (x1 + label_size[0][0], y1 - label_size[0][1]), color, cv2.FILLED)
                        cv2.putText(img_BGR, label_content, (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

        queue.append(new_boxes)

        cv2.imshow("Prediction", imutils.resize(img_BGR, width=800))

        if config["save_result"]:
            VIDEO_OUT.write(img_BGR)

    if cv2.waitKey(1) & 0xFF == ord('s'):
        pause = not pause
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

VIDEO_IN.release()
cv2.destroyAllWindows()
```
